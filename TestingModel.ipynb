#Testing Model

```python
image=tf.keras.preprocessing.image.load_img('/content/drive/MyDrive/Fruit_Vegetable_Recognition/test/cabbage/Image_2.jpg',target_size=(64,64))
input_arr=tf.keras.preprocessing.image.img_to_array(image)
input_arr=np.array([input_arr])
predictions=cnn.predict(input_arr)
```

```python
print(max(predictions[0]))
```

```python
test_set=tf.keras.utils.image_dataset_from_directory(
    '/content/drive/MyDrive/Fruit_Vegetable_Recognition/test',
    labels='inferred',
    label_mode='categorical',
    class_names=None,
    color_mode='rgb',
    batch_size=32,
    image_size=(64,64),
    shuffle=True,
    seed=None,
    validation_split=None,
    subset=None,
    interpolation='bilinear',
    follow_links=False,
    crop_to_aspect_ratio=False
)
```

```python
test_set.class_names
```

```python
result_index=np.where(predictions[0] == max(predictions[0]))
print(result_index[0][0])
```
